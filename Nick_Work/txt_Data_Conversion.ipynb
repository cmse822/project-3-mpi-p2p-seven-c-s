```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_names = [4,8,16,32,64,128,256,512,1024,2048,4096]

data_dict = {}
for i in range(len(data_names)):
    data = pd.read_csv('ping_pong_data/pp_NB_data'+str(data_names[i])+'.txt',skiprows=5,delimiter=',',header=None,index_col=0)
    data_dict[str(data_names[i])+' Bytes'] = np.round(np.mean(data[1]),6)
```

```python
data_dict
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12,12))
plt.scatter(data_names,data_dict.values())
plt.plot(data_names,data_dict.values())

plt.xticks(rotation=90);
plt.yscale('log')
plt.xscale('log')
plt.ylabel('Time (sec)')
plt.xlabel('Bytes Per Message')
plt.title('Ping Pong Non-Blocking Time vs Msg Size')
```
